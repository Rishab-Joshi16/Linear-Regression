```python
X=NHL_R_Stats[['goals_for']]
Y=NHL_R_Stats[['win_pct']]
```

```python
class Linear_Regression():
    def __init__(self, learning_rate, no_of_itr):
        self.learning_rate = learning_rate
        self.no_of_itr = no_of_itr

    def fit(self, x, y):
        self.m, self.n = x.shape
        self.w = np.zeros((self.n, 1))
        self.b = 0
        self.x = x
        self.y = y

        for i in range(self.no_of_itr):
            self.update_weights()

    def update_weights(self):
        y_prediction = self.predict(self.x)
        dw = -(2 * self.x.T).dot(self.y - y_prediction) / self.m
        db = -np.sum(self.y - y_prediction) / self.m
        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db

    def predict(self, x):
        return x.dot(self.w) + self.b

    def return_weights(self):
        print("Weights:", self.w)
        print("Bias:", self.b)
```

```python
x_train = NHL_R_Stats[['goals_for']]
y_train = NHL_R_Stats[[ 'win_pct']]

model = Linear_Regression(learning_rate=0.03, no_of_itr=2000)
model.fit(x_train, y_train)
```
